```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "Date": pd.date_range(start="2024-01-01", periods=10, freq="D"),
    "Actual_Price": [1820, 1830, 1810, 1840, 1855, 1860, 1845, 1870, 1880, 1890],
    "ARIMA_Predicted": [1815, 1822, 1818, 1830, 1840, 1850, 1840, 1855, 1865, 1870],
    "LSTM_Predicted": [1818, 1828, 1812, 1838, 1850, 1858, 1848, 1868, 1878, 1885]
})

df
```

```python
# ARIMA metrics
arima_mae = mean_absolute_error(df["Actual_Price"], df["ARIMA_Predicted"])
arima_rmse = np.sqrt(mean_squared_error(df["Actual_Price"], df["ARIMA_Predicted"]))
arima_r2 = r2_score(df["Actual_Price"], df["ARIMA_Predicted"])

# LSTM metrics
lstm_mae = mean_absolute_error(df["Actual_Price"], df["LSTM_Predicted"])
lstm_rmse = np.sqrt(mean_squared_error(df["Actual_Price"], df["LSTM_Predicted"]))
lstm_r2 = r2_score(df["Actual_Price"], df["LSTM_Predicted"])

print("SPRINT-4 VALIDATION RESULTS")
print("--------------------------------")
print(f"ARIMA → MAE: {arima_mae:.2f}, RMSE: {arima_rmse:.2f}, R²: {arima_r2:.2f}")
print(f"LSTM  → MAE: {lstm_mae:.2f}, RMSE: {lstm_rmse:.2f}, R²: {lstm_r2:.2f}")
```

```python
plt.figure(figsize=(12,5))

plt.plot(df["Date"], df["Actual_Price"], label="Actual Price", linewidth=3)
plt.plot(df["Date"], df["ARIMA_Predicted"], label="ARIMA Prediction", linestyle="--")
plt.plot(df["Date"], df["LSTM_Predicted"], label="LSTM Prediction", linestyle=":")

plt.title("Sprint-4 Validation: Actual vs ARIMA vs LSTM")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
models = ["ARIMA", "LSTM"]
mae = [arima_mae, lstm_mae]
rmse = [arima_rmse, lstm_rmse]
r2 = [arima_r2, lstm_r2]

x = np.arange(len(models))
width = 0.25

plt.figure(figsize=(10,5))
plt.bar(x - width, mae, width, label="MAE")
plt.bar(x, rmse, width, label="RMSE")
plt.bar(x + width, r2, width, label="R²")

plt.xticks(x, models)
plt.ylabel("Metric Value")
plt.title("Sprint-4 Model Performance Comparison")
plt.legend()
plt.grid(axis="y")
plt.show()
```
